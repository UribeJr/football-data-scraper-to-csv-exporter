```python
#import modules and packages
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
```

```python
#scrape single game shots
base_url = 'https://understat.com/match/'
match = str(input("Enter your match ID: "))
url = base_url + match
```

```python
res = requests.get(url)
soup = BeautifulSoup(res.content, 'lxml')
span = soup.find('span')
script = soup.find_all('script')
script
```

```python
string = script[1].string
string
```

```python
#strip symbols so we only have json data
index_start = string.index("('") + 2
index_end = string.index("')")

json_data = string[index_start:index_end]
json_data = json_data.encode('utf8').decode('unicode_escape')
data = json.loads(json_data)
```

```python
#convert each team into separate dataframes
df_h = pd.DataFrame(data['h'])
df_a = pd.DataFrame(data['a'])
```

```python
# Save the home and awayteam DataFrame to a CSV file
df_h.to_csv('home_team_shots.csv', index=False)
df_a.to_csv('away_team_shots.csv', index=False)
```
